# Import Required Libraries
Import the necessary libraries, including NumPy, Matplotlib, and Scikit-learn.

```python
# Import the necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Generate Sample Data
Use Scikit-learn's make_blobs function to generate sample data for clustering.

```python
from sklearn.datasets import make_blobs

# Generate sample data
n_samples = 300
n_features = 2
n_clusters = 4
random_state = 42

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=random_state)

# Plot the sample data
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='k', s=50)
plt.title('Sample Data for K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Visualize Sample Data
Use Matplotlib to visualize the generated sample data.

```python
# Visualize Sample Data
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='o', edgecolor='k', s=50)
plt.title('Sample Data for K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Apply K-Means Clustering
Use Scikit-learn's KMeans class to apply k-means clustering to the sample data.

```python
# Apply K-Means Clustering

# Initialize the KMeans class with the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)

# Fit the KMeans model to the sample data
kmeans.fit(X)

# Predict the cluster labels for the sample data
y_kmeans = kmeans.predict(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.show()
```

# Visualize Clustering Results
Use Matplotlib to visualize the clustering results, including cluster centers and assigned labels.

```python
# Visualize Clustering Results

# Plot the clustered data with assigned labels
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', marker='o', edgecolor='k', s=50)
plt.title('K-Means Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.legend()
plt.show()
```

# Evaluate Clustering Performance
Discuss metrics such as inertia and silhouette score to evaluate the performance of the clustering.

```python
# Evaluate Clustering Performance

# Import necessary libraries for evaluation
from sklearn.metrics import silhouette_score

# Calculate inertia (sum of squared distances to the closest cluster center)
inertia = kmeans.inertia_
print(f'Inertia: {inertia}')

# Calculate silhouette score (measure of how similar an object is to its own cluster compared to other clusters)
silhouette_avg = silhouette_score(X, y_kmeans)
print(f'Silhouette Score: {silhouette_avg}')
```
